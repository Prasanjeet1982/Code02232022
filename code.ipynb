```python
# Import the necessary modules

import pandas as pd
import numpy as np
import json

```

```python
# Opening JSON file
f = open('/content/sample_data/Input.json')
 
# Convert JSON object as a dictionary
data = json.load(f)

# Closing file
f.close()
```

```python
# Verify that the file was read correctly
print(data)
```

```python
type(data)
```

```python
# Convert List to pandas dataframe for easiler calculations
df = pd.DataFrame(data)
df.columns
```

```python
# Calculate BMI with the formula provided
# BMI(kg/m2) = mass(kg) / height(m)2

df['BMI'] = (df['WeightKg']) / ((df['HeightCm']/100)**2)
df
```

```python
# Create a new Lists
BMI_Cat = []
BMI_Range = []
HealthRisk = []

for items in df['BMI']:
    if items <= 18.4:
       BMI_Cat.append('Under Weight')
       BMI_Range.append('18.4 and below')
       HealthRisk.append('Malnutrition risk')
    elif 18.5 < items <= 24.9:
       BMI_Cat.append('Normal Weight')
       BMI_Range.append('18.5 - 24.9')
       HealthRisk.append('Low risk')
    elif 25 < items <= 29.9:
       BMI_Cat.append('Obese')
       BMI_Range.append('25 - 29.9')
       HealthRisk.append('Enhanced risk')
    elif 30 < items <= 34.9:
       BMI_Cat.append('Moderately Obese')
       BMI_Range.append('30 - 34.9')
       HealthRisk.append('Medium risk')
    elif 35 < items <= 39.9:
       BMI_Cat.append('Severly Obese')
       BMI_Range.append('35 - 39.9')
       HealthRisk.append('High risk')
    elif 40 < items:
       BMI_Cat.append('Very Severly Obese')
       BMI_Range.append('40 and above')
       HealthRisk.append('Very high risk')
```

```python
# Add New Columns to Existing data frame
df['BMI_Cat'] = BMI_Cat
df['BMI_Range'] = BMI_Range
df['HealthRisk'] = HealthRisk

df
```
